```python
# Simpan aplikasi streamlit ke mall_apps.py
%%writefile mall_apps.py
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

@st.cache_data
def load_data():
    # Memuat dataset lokal
    df = pd.read_csv("Mall_Customers.csv")
    return df

def main():
    st.set_page_config(page_title="Customer Segmentation", layout="centered")
    st.title("🛍️ Aplikasi Segmentasi Pelanggan Mall")
    st.write("Aplikasi ini mengelompokkan pelanggan berdasarkan Pendapatan Tahunan dan Skor Pengeluaran.")

    df = load_data()

    # Sidebar untuk pengaturan Cluster
    st.sidebar.header("Pengaturan Algoritma")
    k_value = st.sidebar.slider("Pilih jumlah Cluster (K)", 2, 10, 5)

    # Preprocessing sederhana (mengambil kolom Annual Income dan Spending Score)
    X = df.iloc[:, [3, 4]].values

    # Model K-Means
    kmeans = KMeans(n_clusters=k_value, init='k-means++', random_state=42)
    y_kmeans = kmeans.fit_predict(X)
    df['Cluster'] = y_kmeans

    # Input User untuk Prediksi Cluster Baru
    st.subheader("Input Data Pelanggan Baru")
    col1, col2 = st.columns(2)
    with col1:
        income = st.number_input("Annual Income (k$)", min_value=0, max_value=150, value=50)
    with col2:
        score = st.number_input("Spending Score (1-100)", min_value=1, max_value=100, value=50)

    if st.button("Cek Cluster Pelanggan"):
        user_data = np.array([[income, score]])
        prediction = kmeans.predict(user_data)[0]
        st.success(f"Pelanggan ini masuk ke dalam **Cluster {prediction}**")

    # Visualisasi
    st.subheader("Visualisasi Cluster")
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_kmeans, palette='viridis', s=100, ax=ax)
    ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids', marker='X')
    plt.xlabel('Annual Income (k$)')
    plt.ylabel('Spending Score (1-100)')
    plt.legend()
    st.pyplot(fig)

    if st.checkbox("Tampilkan Data Mentah"):
        st.dataframe(df)

if __name__ == "__main__":
    main()
```
